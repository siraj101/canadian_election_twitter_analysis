# Preprocessing for Topic Modeling

```python
df
```

```python
import nltk
```

```python
import random
```

```python
nltk.download('stopwords')
```

```python
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer
```

# Removing hyperlinks and twitter-specific styles

```python
for i in range(len(df.renderedContent)):
    df.renderedContent[i] = re.sub(r'^RT[\s]+', '', df.renderedContent[i])
```

```python
for i in range(len(df.renderedContent)):
    df.renderedContent[i] = re.sub(r'https?:\/\/.*[\r\n]*', '', df.renderedContent[i])
```

```python
for i in range(len(df.renderedContent)):
    df.renderedContent[i] = re.sub(r'#', '', df.renderedContent[i])
```

```python
for i in range(len(df.renderedContent)):
    df.renderedContent[i] = re.sub(r'[0-9]', '', df.renderedContent[i])
```

```python
df.renderedContent
```

# Tokenization

```python
tokenizer = TweetTokenizer(preserve_case = False, strip_handles = True, reduce_len = True)
```

```python
for i in range(len(df.renderedContent)):
    df.renderedContent[i] = tokenizer.tokenize(df.renderedContent[i])
```

```python
df.renderedContent
```

# Remove stopwords and punctuation

```python
stopwords_english = stopwords.words('english')
```

```python
clean_tweets = []

for i in range(len(df.renderedContent)):
    for j in df.renderedContent[i]:
        if (j not in stopwords_english and j not in string.punctuation):
            clean_tweets.append(j)
clean_tweets
```

# Stemming

```python
stemmer = PorterStemmer()
```

```python
stem_tweets = []

for i in clean_tweets:
    stem_word = stemmer.stem(i)
    stem_tweets.append(stem_word)

stem_tweets
```

# Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = stem_tweets
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit_transform(corpus)
```

```python
vectorizer.vocabulary_
```

```python
bagofwords_df = pd.DataFrame(vectorizer.vocabulary_, index = [i for i in range(len(vectorizer.vocabulary_))])
```

```python
bdf = bagofwords_df.iloc[0, :]
```

```python
bdf = pd.DataFrame(bdf)
bdf
```

```python
bdf = bdf.rename(columns = {0: 'Count'})
bdf
```

```python
bdf = bdf.reset_index()
```

```python
bdf = bdf.rename(columns = {'index': 'word'})
bdf
```
